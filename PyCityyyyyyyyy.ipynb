```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to load
students_complete_df = pd.read_csv('students_complete.csv')
schools_complete_df = pd.read_csv('schools_complete.csv')
```

```python
# Read Students data
'students_complete_df'
```

```python
# Read School data
schools_complete_df
```

```python
# Merging data into single Dataset
combined_df = pd.merge(students_complete_df,schools_complete_df, how="outer", on="school_name")
```

```python
combined_df
```

```python
# Calculating the total number of schools 
schools_count = len(combined_df["school_name"].unique())
schools_count
```

```python
#Calculating the total number of students
total_student_count = combined_df["student_name"].count()
total_student_count
```

```python
# Calculating the total budget
total_budget = sum(combined_df["budget"].unique())
total_budget
```

```python
# Calculating the average math score
average_mathscore = combined_df["math_score"].mean()
average_mathscore
```

```python
# Calculating the average reading score
average_readingscore= combined_df["reading_score"].mean()
average_readingscore
```

```python
# Calculating the percentage of students with a passing math score (70 or greater)
percent_mathscore= combined_df.query('reading_score >= 70').math_score.count() / combined_df["reading_score"].count() * 100
percent_mathscore
```

```python
# Calculating the percentage of students with a passing reading score (70 or greater)
percent_readingscore= combined_df.query('math_score >= 70').math_score.count() / combined_df["math_score"].count() * 100
percent_readingscore
```

```python
total_combined_per = combined_df[(combined_df["reading_score"]>= 70) & (combined_df["math_score"]>=70)].count()["student_name"]/total_student_count * 100
total_combined_per
```

```python
# Creating a Dataframe to hold the above results

df = pd.DataFrame([{"Total Scchool":schools_count,"Total Students":total_student_count, 
                   "Total Budget":total_budget, "Average Math Score":average_mathscore,
                    "Average Reading Score":average_readingscore, "% Passing Math":percent_mathscore,
                   "% Passing Reading":percent_readingscore, "% Overall Passing":total_combined_per
                   
                   
                   }])
df
```

```python
# Average Math Score
per_school_math = combined_df.groupby(["school_name"]).mean()["math_score"]
per_school_math
```

```python
#total passing scores

per_school_math_passing = combined_df[combined_df["math_score"]>=70].groupby("school_name").count()["math_score"]
per_school_math_passing
```

```python
per_school_reading_passing = combined_df[combined_df["reading_score"]>=70].groupby("school_name").count()["reading_score"]
per_school_reading_passing
```

```python
# Average Reading Score
avg_reading= combined_df.groupby(["school_name"]).mean()["reading_score"]
avg_reading
```

```python
# % Passing Math
percent_math= per_school_math_passing/count_by_school *100
percent_math
```

```python
# % Passing Reading
percent_reading = per_school_reading_passing/count_by_school *100
percent_reading
```

```python
overall = (combined_df["reading_score"]>= 70) & (combined_df["math_score"]>=70)
passed_overall_mark = combined_df[overall]
overall_mark_per_school = passed_overall_mark.groupby(["school_name"]).count()["Student ID"]
overall_mark_per_school
```

```python
# % Overall Passing
percent_overall = (overall_mark_per_school/count_by_school)*100
percent_overall 
```
